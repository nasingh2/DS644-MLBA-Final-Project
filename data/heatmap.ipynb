### Heatmap of Zillow Properties

```python
import pandas as pd 

# importing final cleaned dataset
df = pd.read_csv('dataset_final.csv')

# printing first few rows 
print(df.head())
```

```python
import folium
from folium.plugins import HeatMap
import branca

# creating map values 
location_price = pd.DataFrame(df, columns = ['latitude', 'longitude', 'price'])

# dropping missing and null values 
location_price = location_price.dropna()

# checking how many missing values from df now 
# location_price.isna().sum()

# creating map
m = folium.Map(location = [df.latitude.mean(), df.longitude.mean()], zoom_start = 6, control_scale = True)

# creating heatmap layers
data = location_price.values.tolist()

# making heatmap 
heat_map = HeatMap(data, min_opacity = 0.3, max_opacity = 0.5, radius = 25).add_to(m)

# showing map 
m

# saving map to html file 
# m.save('properties_price_heatmap.html')
```
